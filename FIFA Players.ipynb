```python
# weight cleaning
data['Weight'] = data['Weight'].str.replace('lbs', '', regex=False)
data['Weight'] = data['Weight'].astype(int)
```

```python
# height cleaning
def height_to_float(height):
    feet, inches = height.split("'")
    inches = inches.replace('"', '')
    feet = int(feet)
    inches = int(inches)
    return feet + inches/12
data['Height'] = data['Height'].apply(height_to_float)
data['Height']
```

```python
# create 3 separated columns of year, month, date for 'joined' column

data['Joined'] = pd.to_datetime(data['Joined'])
data['Year'] = data['Joined'].dt.year
data['Month'] = data['Joined'].dt.month
data['Day'] = data['Joined'].dt.day
data[['Joined', 'Year', 'Month', 'Day']]
```

```python
# converting money column to inteegr
def money_to_integer(money):
    money = str(money)
    money = money.replace('€', '')
    lastc = money[-1]
    if lastc == 'M':
        money = money.replace('M', '')
        money = float(money)
        money = money * 1000000
    elif lastc == 'K':
        money = money.replace('K', '')
        money = float(money)
        money = money * 1000
    return int(money)
data['Value'] = data['Value'].apply(money_to_integer)
data['Wage'] = data['Wage'].apply(money_to_integer)
data['Release Clause'] = data['Release Clause'].apply(money_to_integer)
data[['Value', 'Wage', 'Release Clause']]
```

```python
data['W/F'] = data['W/F'].str.replace('★', '')
data['SM'] = data['SM'].str.replace('★', '')
data['IR'] = data['IR'].str.replace('★', '')
data[['W/F', 'SM', 'IR']]
```

```python
# clean hits
def string_to_array(hits):
    hits = str(hits).replace('\n', '')
#     hits = hits.split(' ')
    return hits
data['Hits'] = data['Hits'].apply(string_to_array)

data['Hits']
```

```python
correlation = data.corr(method='spearman')
# Define a threshold for "good" correlation
threshold = -0.7

# Stack the correlation matrix, reset the index, and filter
strong_correlations = (
    correlation
    .stack()
    .reset_index()
    .query("level_0 < level_1")
    .sort_values(by=0, ascending=False)
)

# Rename columns for readability
strong_correlations.columns = ['Column1', 'Column2', 'Correlation']

print(strong_correlations)
```

```python
filtered = strong_correlations[strong_correlations['Correlation'].astype(float) > 0.8]
filtered
```

```python

def grouping(value1, value2, groups):
    
    for group in groups:
        if value1 in group:
            group.update([value1, value2])
            return
        elif value2 in group:
            group.update([value1, value2])
            return
    # If not found in any group, create a new one
    groups.append(set([value1, value2]))
groups = []
for i, row in filtered.iterrows():
    grouping(row['Column1'], row['Column2'], groups)

groups
```
